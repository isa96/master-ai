# Membership Value

```python
def get_mu_as(v_x):
    mask_1 = (0 <= v_x) & (v_x <= 0.5)
    mask_x = (0.5 < v_x) & (v_x <= 0.75)
    mask_0 = (0.75 < v_x) & (v_x <= 1)
    
    mu = v_x.copy()
    mu[mask_1] = 1
    mu[mask_x] = 3 - 4 * v_x[mask_x]
    mu[mask_0] = 0

    return mu

def get_mu_hs(v_x):
    mask_0 = (0 <= v_x) & (v_x <= 0.5)
    mask_x = (0.5 < v_x) & (v_x <= 0.75)
    mask_1 = (0.75 < v_x) & (v_x <= 1)
    
    mu = v_x.copy()
    mu[mask_0] = 0
    mu[mask_x] = 4 * v_x[mask_x] - 2
    mu[mask_1] = 1

    return mu

def get_mu_lf(f_x_y):
    mu = f_x_y.copy()
    mask_0 = (0.5 <= f_x_y) & (f_x_y <= 1.0)
    mask_x = (0.25 <= f_x_y) & (f_x_y <= 0.5)
    mask_1 = (0.0 <= f_x_y) & (f_x_y <= 0.25)

    mu[mask_0] = 0
    mu[mask_x] = 2 - 4 * f_x_y
    mu[mask_1] = 1

    return mu

def get_mu_ff(f_x_y):
    mu = f_x_y.copy()
    mask_x1 = (0.0 <= f_x_y) & (f_x_y <= 0.25)
    mask_1 = (0.25 <= f_x_y) & (f_x_y <= 0.5)
    mask_x2 = (0.5 <= f_x_y) & (f_x_y <= 0.75)
    mask_0 = (0.75 <= f_x_y) & (f_x_y <= 1.0)

    mu[mask_x1] = 4 * f_x_y
    mu[mask_1] = 1
    mu[mask_x2] = 3 - 4 * f_x_y
    mu[mask_0] = 0

    return mu

def get_mu_hf(f_x_y):
    mu = f_x_y.copy()
    mask_0 = (0 <= f_x_y) & (f_x_y <= 0.25)
    mask_x = (0.25 <= f_x_y) & (f_x_y <= 0.5)
    mask_1 = (0.5 <= f_x_y) & (f_x_y <= 1.0)

    mu[mask_0] = 0
    mu[mask_x] = 4 * f_x_y - 1
    mu[mask_1] = 1

    return mu

def get_mu_rl(d_x_y):
    mu = d_x_y.copy()
    mask_0 = (0 <= d_x_y) & (d_x_y <= 0.5)
    mask_x = (0.5 <= d_x_y) & (d_x_y <= 0.75)
    mask_1 = (0.75 <= d_x_y) & (d_x_y <= 1.0)

    mu[mask_0] = 0
    mu[mask_x] = 4 * d_x_y - 2
    mu[mask_1] = 1

    return mu

def get_mu_rf(d_x_y):
    mu = d_x_y.copy()
    mask_1 = (0 <= d_x_y) & (d_x_y <= 0.5)
    mask_x = (0.5 <= d_x_y) & (d_x_y <= 0.75)
    mask_0 = (0.75 <= d_x_y) & (d_x_y <= 1.0)

    mu[mask_0] = 0
    mu[mask_x] = 3 - 4 * d_x_y
    mu[mask_1] = 1


    return mu
```

```python
mu_as = get_mu_as(skill)
mu_hs = get_mu_hs(skill)
mu_lf = get_mu_lf(freq)
mu_ff = get_mu_ff(freq)
mu_hf = get_mu_hf(freq)
mu_rl = get_mu_rl(recom)
mu_rf = get_mu_rf(recom)

mu_rf
```

```python
x_skil = np.arange(0.0, 1.05, 0.05)
x_group_frequency = np.arange(0.0, 1.05, 0.05)
x_group_recomendation = np.arange(0, 1.05, 0.05)

# SKILL
mu_AS = fuzz.trapmf(x_skil, [0.0, 0.0, 0.50, 0.75])
mu_HS = fuzz.trapmf(x_skil, [0.50, 0.75, 1.0, 1.0])

# Group Frequency
mu_LF = fuzz.trapmf(x_group_frequency, [0.0, 0.0, 0.25, 0.50])
mu_FF = fuzz.trapmf(x_group_frequency, [ 0.0, 0.25, 0.50, 0.75])
mu_HF = fuzz.trapmf(x_group_frequency, [0.25, 0.50, 1.0, 1.0])

# Group Recomendation
mu_RL = fuzz.trapmf(x_group_recomendation, [0.50, 0.75, 1.0, 1.0 ])
mu_RF = fuzz.trapmf(x_group_recomendation, [0.0, 0.0, 0.50, 0.75])

fig, (ax0, ax1, ax2) = plt.subplots(nrows = 3, figsize=(10,5))
ax0.plot(x_skil, mu_AS, 'g', linewidth=1.5, label='AS')
ax0.plot(x_skil, mu_HS, 'r', linewidth=1.5, label='HS')

ax1.plot(x_group_frequency, mu_LF, 'g', linewidth=1.5, label='LF')
ax1.plot(x_group_frequency, mu_FF, 'r', linewidth=1.5, label='FF')
ax1.plot(x_group_frequency, mu_HF, 'b', linewidth=1.5, label='HF')

ax2.plot(x_group_recomendation, mu_RL, 'g', linewidth=1.5, label='RL')
ax2.plot(x_group_recomendation, mu_RF, 'r', linewidth=1.5, label='RF')

for ax in (ax0, ax1, ax2):
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.set_xticks([0., 0.25, 0.5, 0.75, 1.0])
  ax.grid(linestyle='--')
  ax.set_axisbelow(True)
  ax.get_xaxis().tick_bottom()
  ax.get_yaxis().tick_left()
  ax.legend()

ax0.set_title('Skill (Average & High)')
ax1.set_title('Frequency (Lowly | Fairly | Highly)')
ax2.set_title('Recomendation (Less & Fairly)')

plt.tight_layout()
```

# Simulation

```python
def standard_t_norm(x, y):
    return min(x, y)

def algebraic_t_norm(x, y):
    return x * y

def lukasiewicz_t_norm(x, y):
    return max(0, x + y - 1)

def get_yager_t_norm(ld=0.5):
    def yager_t_norm(x, y):
        return 1. - min(1.,
            np.power(np.power((1 - x), ld) + np.power((1 - y), ld), 1. / ld)
        )
    return yager_t_norm

def hamacher_t_norm(x, y):
    if x == 0 and y == 0:
        return 0
    return (x * y) / (x + y - x * y)
```

```python
pair = 'A, C'
x, y = pair.split(', ')

t_norm = get_yager_t_norm(ld=0.75)

less_recom_matrix = pd.DataFrame(index=skill.index, columns=skill.index)
less_recom_matrix = less_recom_matrix.fillna(0)

fair_recom_matrix = pd.DataFrame(index=skill.index, columns=skill.index)
fair_recom_matrix = fair_recom_matrix.fillna(0)

for x in skill.index:
    for y in skill.index:
        if x == y:
            continue

        less_recom = max(
            t_norm(mu_hs[x], mu_hs[y]), # rule 1
            mu_hf.loc[x, y] # rule 5
        )

        fair_recom = max(
            t_norm(mu_as[x], mu_hs[y]), # rule 2
            t_norm(mu_hs[x], mu_as[y]), # rule 3
            t_norm(mu_as[x], mu_as[y]), # rule 4
            mu_ff.loc[x, y], # rule 6
            mu_lf.loc[x, y] # rule 7
        )

        less_recom_matrix.loc[x, y] = less_recom
        fair_recom_matrix.loc[x, y] = fair_recom
```
